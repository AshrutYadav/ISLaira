```python
import cv2
import ipywidgets as widgets
from IPython.display import display

# 1. Create the persistent "TV screen" widget
# This is the "overlap" box that stays in one place
image_widget = widgets.Image(format='jpeg', width=640, height=480)
display(image_widget)

cap = cv2.VideoCapture("http://100.127.249.15:8080/video") # or your Tailscale URL

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # 2. Add your MediaPipe/Processing here
        # (Draw on 'frame' directly using cv2)

        # 3. Encode to JPEG (The "Signal" for the TV)
        _, buffer = cv2.imencode('.jpg', frame)
        
        # 4. SWAP the value (This is the "Overlap")
        image_widget.value = buffer.tobytes()
        display()
except KeyboardInterrupt:
    cap.release()
```
